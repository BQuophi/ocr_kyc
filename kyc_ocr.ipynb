## Text Recognition


```python
from paddleocr import PaddleOCR
ocr = PaddleOCR() # need to run only once to download and load model into memory

img_path = "voter's_card_3.jpg"
result = ocr.ocr(img_path, det=True, rec=True)

results = []

for line in result:
  results.append(line[0][1]) # Country
  results.append(line[1][1]) # Electoral Body
  results.append(line[2][1]) # ID Type
  results.append(line[3][1]) # Code 
  results.append(line[4][1])
  results.append(line[5][1])
  results.append(line[6][1]) # Name
  results.append(line[7][1]) # Date Of Birth
  results.append(line[8][1])
  results.append(line[9][1]) # Gender
  results.append(line[10][1]) # Occupation
  results.append(line[11][1])
  results.append(line[12][1]) # Address
  results.append(line[13][1]) # Address
  results.append(line[14][1])
  results.append(line[15][1])
  results.append(line[16][1])
  results.append(line[17][1])
  results.append(line[18][1])
  results.append(line[19][1])
  results.append(line[20][1])
 
```
